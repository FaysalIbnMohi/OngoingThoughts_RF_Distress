```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

# Load the dataset
file_path = "/Users/faisalmahmud/Desktop/RF_Classification_to_classify_DASS_from_Daily_Ongoing_Thoughts/Binary/Preprocessing/Corrected_Unscaled_thought_personality_combined.csv"
data = pd.read_csv(file_path)

# Classification functions based on the provided thresholds
def classify_depression(value):
    if 0 <= value <= 9 or 10 <= value <= 13 or 14 <= value <= 20:
        return 0  # Normal + Mild + Moderate
    else:
        return 1  # Severe + Extremely Severe

def classify_anxiety(value):
    if 0 <= value <= 7 or 8 <= value <= 9 or 10 <= value <= 14:
        return 0  # Normal + Mild + Moderate
    else:
        return 1  # Severe + Extremely Severe

def classify_stress(value):
    if 0 <= value <= 14 or 15 <= value <= 18 or 19 <= value <= 25:
        return 0  # Normal + Mild + Moderate
    else:
        return 1  # Severe + Extremely Severe

def classify_dass_total(value):
    # Calculate the sum of the upper bounds for the "Normal" categories
    normal_threshold = 9 + 7 + 14  # Sum of Normal max values
    moderate_threshold = 27 + 19 + 33  # Sum of Moderate max values
    
    if value <= normal_threshold:
        return 0  # Normal + Mild + Moderate
    else:
        return 1  # Severe + Extremely Severe

# Apply classification functions to create new categorical columns
data['DASS_D_Class'] = data['DASS_D'].apply(classify_depression)
data['DASS_A_Class'] = data['DASS_A'].apply(classify_anxiety)
data['DASS_S_Class'] = data['DASS_S'].apply(classify_stress)
data['DASS_total_Class'] = data['DASS_total'].apply(classify_dass_total)

# Count total instances for each class BEFORE splitting
print("\nTotal instances per class BEFORE splitting:")
print("DASS_total_Class:", data['DASS_total_Class'].value_counts())

# Perform an 80-20 split while keeping near balance
train_data, test_data = train_test_split(
    data, test_size=0.2, stratify=data[['DASS_total_Class']], random_state=42
)

# Count instances for each class AFTER splitting
print("\nTotal instances per class in TRAIN dataset:")
print("DASS_total_Class:\n", train_data['DASS_total_Class'].value_counts())

print("\nTotal instances per class in TEST dataset:")
print("DASS_total_Class:\n", test_data['DASS_total_Class'].value_counts())

# Define output directories for saving the datasets
train_output_dir = "/Users/faisalmahmud/Desktop/RF_Classification_to_classify_DASS_from_Daily_Ongoing_Thoughts/Binary/FinalData"
test_output_dir = train_output_dir  # Same directory for test data

# Check if the directory exists, if not, create it
if not os.path.exists(train_output_dir):
    os.makedirs(train_output_dir)

# Save the train and test datasets
train_output_file = os.path.join(train_output_dir, "DASS_TrainData.csv")
test_output_file = os.path.join(test_output_dir, "DASS_TestData.csv")

train_data.to_csv(train_output_file, index=False)
test_data.to_csv(test_output_file, index=False)

# Display final dataset info
print(f"\nTrain dataset saved to: {train_output_file} (Size: {len(train_data)})")
print(f"Test dataset saved to: {test_output_file} (Size: {len(test_data)})")

# Plotting bar charts for the distribution of class 0 and class 1 (Normal+Mild+Moderate vs Severe+Extremely Severe)
def plot_class_distribution(data, title):
    plt.figure(figsize=(8, 6))
    sns.countplot(x=data, palette='Set2')
    plt.title(title)
    plt.xlabel('Class (0: Normal + Mild + Moderate, 1: Severe + Extremely Severe)')
    plt.ylabel('Count')
    plt.xticks([0, 1], ['0: Normal + Mild + Moderate', '1: Severe + Extremely Severe'])
    plt.show()

# Plot bar charts for train and test datasets
plot_class_distribution(train_data['DASS_total_Class'], 'Train Dataset Class Distribution')
plot_class_distribution(test_data['DASS_total_Class'], 'Test Dataset Class Distribution')
```
