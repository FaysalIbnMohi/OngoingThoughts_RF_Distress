```python
# ----------------------------------------------
# üì¶ Import Required Libraries
# ----------------------------------------------
import os
import shap
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import (
    accuracy_score, confusion_matrix, classification_report,
    precision_score, recall_score, f1_score, roc_curve, roc_auc_score
)
from sklearn.preprocessing import label_binarize
from imblearn.over_sampling import SMOTE

# ----------------------------------------------
# üìÇ Load Data
# ----------------------------------------------
train_path = "/Users/faisalmahmud/Library/CloudStorage/OneDrive-UniversityofPlymouth/PhD/Chapter/P2_reallife/DASS/FinalDataAnqi/FinalData/DASS_TrainData_copy.csv"
test_path = "/Users/faisalmahmud/Library/CloudStorage/OneDrive-UniversityofPlymouth/PhD/Chapter/P2_reallife/DASS/FinalDataAnqi/FinalData/DASS_TestData_copy.csv"

df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

# ----------------------------------------------
# üîß Configuration
# ----------------------------------------------
features = [
    'Task', 'Future', 'Past', 'Self', 'People', 'External', 'Images', 'Words',
    'Sounds', 'Detailed', 'Deliberate', 'Solution', 'Intrusive', 'Knowledge',
    'Absorption', 'Distracting', 'Meaningful'
]

target_variables = ['DASS_totalM_Class', 'DASS_DM_Class', 'DASS_AM_Class', 'DASS_SM_Class']

label_map = {
    'DASS_totalM_Class': 'DASS Total',
    'DASS_DM_Class': 'Depression',
    'DASS_AM_Class': 'Anxiety',
    'DASS_SM_Class': 'Stress'
}

class_names = ['Mild', 'Moderate', 'Severe']  # 0, 1, 2
combined_output_dir = "/Users/faisalmahmud/Library/CloudStorage/OneDrive-UniversityofPlymouth/PhD/Chapter/P2_reallife/DASS/FinalDataAnqi/Results/CombinedTargets/TogetherFinalMulticlass"
os.makedirs(combined_output_dir, exist_ok=True)

# ----------------------------------------------
# üöÄ Train Models + Collect Results (Multiclass)
# ----------------------------------------------
results = {}

for target_variable in target_variables:
    print(f"\nüîç Processing target: {target_variable}")
    
    X_train = df_train[features].values
    y_train = df_train[target_variable].values
    X_test = df_test[features].values
    y_test = df_test[target_variable].values

    smote = SMOTE(random_state=42)
    X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

    X_train_split, X_val, y_train_split, y_val = train_test_split(
        X_train_res, y_train_res, test_size=0.2, random_state=42, stratify=y_train_res
    )

    param_grid = {
        'n_estimators': [100, 200],
        'max_depth': [20, 30],
        'max_features': ['sqrt'],
        'min_samples_split': [2],
        'min_samples_leaf': [1],
        'bootstrap': [True]
    }

    rf = RandomForestClassifier(random_state=42)
    grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train_split, y_train_split)
    best_rf = grid_search.best_estimator_

    y_pred = best_rf.predict(X_val)
    y_probs = best_rf.predict_proba(X_val)

    # Evaluation Metrics
    accuracy = accuracy_score(y_val, y_pred)
    precision = precision_score(y_val, y_pred, average='weighted')
    recall = recall_score(y_val, y_pred, average='weighted')
    f1 = f1_score(y_val, y_pred, average='weighted')
    roc_auc = roc_auc_score(label_binarize(y_val, classes=[0, 1, 2]), y_probs, multi_class='ovr')

    print(f"‚úÖ {label_map[target_variable]}")
    print(f"  Accuracy:  {accuracy:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall:    {recall:.4f}")
    print(f"  F1 Score:  {f1:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_val, y_pred, target_names=class_names))

    explainer = shap.TreeExplainer(best_rf)
    shap_values = explainer.shap_values(X_val)

    results[target_variable] = {
        'model': best_rf,
        'X_val': X_val,
        'y_val': y_val,
        'y_pred': y_pred,
        'y_probs': y_probs,
        'confusion_matrix': confusion_matrix(y_val, y_pred),
        'roc_auc': roc_auc,
        'shap_values': shap_values
    }

# ----------------------------------------------
# üìä Plot: Combined Confusion Matrices (2x2)
# ----------------------------------------------
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.ravel()

for i, target_variable in enumerate(target_variables):
    cm = results[target_variable]['confusion_matrix']
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names,
                cbar=False, ax=axes[i])
    axes[i].set_title(f"Confusion Matrix: {label_map[target_variable]}")
    axes[i].set_xlabel("Predicted")
    axes[i].set_ylabel("True")

plt.tight_layout()
plt.savefig(os.path.join(combined_output_dir, "combined_confusion_matrices_multiclass.png"))
plt.show()

# ----------------------------------------------
# üìà Plot: ROC AUC Scores (Bar Chart)
# ----------------------------------------------
plt.figure(figsize=(10, 6))
auc_scores = [results[t]['roc_auc'] for t in target_variables]
labels = [label_map[t] for t in target_variables]
sns.barplot(x=labels, y=auc_scores)
plt.ylabel("ROC AUC (OvR)")
plt.title("Multiclass ROC AUC Scores")
plt.ylim(0, 1)
plt.grid(True)
plt.savefig(os.path.join(combined_output_dir, "roc_auc_multiclass_bar.png"))
plt.show()

# ----------------------------------------------
# üîç Plot: SHAP Summary (All Classes, Individually)
# ----------------------------------------------
for target_variable in target_variables:
    shap_values = results[target_variable]['shap_values']
    X_val = results[target_variable]['X_val']
    
    for class_idx, class_label in enumerate(class_names):
        plt.figure(figsize=(10, 8))
        shap.summary_plot(shap_values[class_idx], X_val, feature_names=features, show=False)
        plt.title(f"{label_map[target_variable]} - Class {class_label}")
        
        filename = f"shap_summary_{target_variable}_class{class_idx}.png"
        plt.savefig(os.path.join(combined_output_dir, filename), bbox_inches='tight')
        plt.show()
        plt.close()

# ----------------------------------------------
# üîç Plot: SHAP Summary (Top 10, Severe class only, Combined 2x2)
# ----------------------------------------------
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for i, target_variable in enumerate(target_variables):
    shap_values = results[target_variable]['shap_values']
    X_val = results[target_variable]['X_val']
    
    plt.sca(axes[i])
    shap.summary_plot(shap_values[2], X_val, feature_names=features, show=False, max_display=10)
    axes[i].set_title(f"{label_map[target_variable]} - Severe")
    axes[i].set_xlabel('')  # Removes SHAP footer

plt.tight_layout()
path = os.path.join(combined_output_dir, "combined_shap_summary_top10_severe_2x2.png")
plt.savefig(path, bbox_inches='tight')
plt.show()
```
