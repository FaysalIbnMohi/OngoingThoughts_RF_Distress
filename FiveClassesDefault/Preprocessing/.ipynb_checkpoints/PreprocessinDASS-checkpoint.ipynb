```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the dataset
file_path = "/Users/faisalmahmud/Library/CloudStorage/OneDrive-UniversityofPlymouth/PhD/Chapter/P2_reallife/DASS/Old/Preprocessing/UpdatedDataset_with_ClinicalTAS.csv"
data = pd.read_csv(file_path)

# Define the column of interest
column = 'TAS_total'

# Categorize TAS_total into ClinicalTAS
def classify_tas(value):
    if value <= 50:
        return 0  # No Alexithymia
    elif 51 <= value <= 60:
        return 1  # Borderline Alexithymia
    else:
        return 2  # Alexithymia

# Apply classification
data['ClinicalTAS'] = data[column].apply(classify_tas)

# Count total instances for each class BEFORE splitting
count_class_0 = len(data[data['ClinicalTAS'] == 0])
count_class_1 = len(data[data['ClinicalTAS'] == 1])
count_class_2 = len(data[data['ClinicalTAS'] == 2])

print("\nTotal instances per class BEFORE splitting:")
print(f"Class 0 (No Alexithymia): {count_class_0}")
print(f"Class 1 (Borderline Alexithymia): {count_class_1}")
print(f"Class 2 (Alexithymia): {count_class_2}")

# Perform an 80-20 split while keeping near balance
train_data, test_data = train_test_split(
    data, test_size=0.2, stratify=data['ClinicalTAS'], random_state=42
)

# Count instances for each class AFTER splitting
train_class_counts = train_data['ClinicalTAS'].value_counts()
test_class_counts = test_data['ClinicalTAS'].value_counts()

print("\nTotal instances per class in TRAIN dataset:")
print(train_class_counts)

print("\nTotal instances per class in TEST dataset:")
print(test_class_counts)

# Save the train and test datasets
train_output_file = "/Users/faisalmahmud/Library/CloudStorage/OneDrive-UniversityofPlymouth/PhD/Chapter/P2_reallife/DASS/Old/FinalData/DASSTrainData.csv"
test_output_file = "/Users/faisalmahmud/Library/CloudStorage/OneDrive-UniversityofPlymouth/PhD/Chapter/P2_reallife/DASS/Old/FinalData/DASSTestData.csv"

train_data.to_csv(train_output_file, index=False)
test_data.to_csv(test_output_file, index=False)

# Display final dataset info
print(f"\nTrain dataset saved to: {train_output_file} (Size: {len(train_data)})")
print(f"Test dataset saved to: {test_output_file} (Size: {len(test_data)})")

# Display class distributions as percentages
print("\nTrain Class Distribution (percentages):\n", train_class_counts / len(train_data) * 100)
print("\nTest Class Distribution (percentages):\n", test_class_counts / len(test_data) * 100)
```
